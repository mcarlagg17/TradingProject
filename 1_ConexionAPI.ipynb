```python
# Librerias
import os 
import requests
import pandas as pd
import json
import dotenv
```

**NOTA:** en este archivo se ha utilizado la `API_KEY` proporcionada en la página de RapidAPI

>Es necesario crear un usuario para conseguirla 


```python
# cargamos las variables de entorno
dotenv.load_dotenv()  

# leemos las variables de entorno
YOUR_API_KEY = os.getenv('API_KEY')
```

```python
url_base = 'https://twelve-data1.p.rapidapi.com/'

url_data_selected = 'time_series'

# Parametros a modificar 
APIparams = {
    'symbol': 'AMZN', # simbolo de la empresa
    'interval': '1day', # intervalo temporal
    'outputsize': '600', # tamaño de la salida
    'format': 'json' # formato de salida
}

header = {
    'X-RapidAPI-Key': YOUR_API_KEY,
    'X-RapidAPI-Host': 'twelve-data1.p.rapidapi.com'
}

url = url_base+url_data_selected
response = requests.get(url,params=APIparams,headers=header)
```

```python
response.json()
```

```python
njson= response.json()  
```

```python
# Separación de la data en meta y los valores
meta_json = njson['meta']
values_json = njson['values']
```

```python
values_json
```

```python
# Normalizado del json, para leerlo en forma de dataframe
df = pd.json_normalize(values_json)
df_meta = pd.json_normalize(meta_json)
```

```python
# Datos guardados en la carpeta data con el nombre correspondiente
file_name= 'time_series_data_01.csv'
df.to_csv('src/data/'+file_name)

#df_meta.to_csv('src/data/meta_'+file_name) # se descomenta si se quiere guardar la información meta
```
